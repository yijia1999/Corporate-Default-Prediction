```python
dsf = pd.read_csv("dsf_new.csv", usecols=['CUSIP', 'PERMNO','DATE', 'PRC', 'SHROUT', 'RET'], parse_dates=['DATE'])
dsf.columns = dsf.columns.str.lower().str.replace(' ', '_')
dsf['ret'] = dsf['ret'].replace('C', np.nan)
dsf = dsf.dropna()
dsf.reset_index(drop=True, inplace=True)
dsf['shrout'] = dsf['shrout']*1000
dsf['ret'] = dsf['ret'].astype('float')
dsf['year'] = pd.DatetimeIndex(dsf['date']).year
dsf['cusip'] = dsf['cusip'].astype('str').str[0:6]
dsf['permno'] = dsf['permno'].astype('str')
dsf['E'] = abs(dsf['prc']) * dsf['shrout']
dsf
```

```python
# annual dsf 
annret = dsf.groupby(by=['cusip', 'permno','year']).apply(lambda x: np.exp(np.sum(np.log(1+x['ret'])))).reset_index()
sigmae = dsf.groupby(by=['cusip', 'permno','year'])['ret'].std()*np.sqrt(250)
E = dsf.groupby(by=['cusip','permno', 'year'])['E'].first()
annual_dsf = pd.DataFrame(annret).merge(sigmae, on=['cusip', 'permno','year']).merge(E, on=['cusip', 'permno','year'])
annual_dsf.rename(columns={0: 'annret', 'ret': 'sigmae'}, inplace=True)
annual_dsf['annret'] = annual_dsf.sort_values(by=['year'], ascending=True).groupby(by=['cusip','permno'])['annret'].shift(1)
annual_dsf['sigmae'] = annual_dsf.sort_values(by=['year'], ascending=True).groupby(by=['cusip'])['sigmae'].shift(1)
annual_dsf.dropna(inplace=True)
annual_dsf.reset_index(drop=True, inplace=True)
annual_dsf
```

```python
# Merge Data: crsp funda bankruptcy 
annual = (annual_dsf.merge(bankruptcy, on = ['year','permno'], how = 'left').merge(funda, on=['year', 'cusip'], how = 'left').merge(dailyfed, on=['year'], how = 'left'))
annual['check_bankruptcy'] = annual['check_bankruptcy'].fillna(0)
annual = annual.drop(columns=['permno'])
annual = annual.sort_values(by=['cusip','year'])
# DD PD 
def KMV(data, a, b):
    df = data.copy()
    T = 1
    df['V'] = df['E'] + df['F']
    df['sigmad'] = a+b*df['sigmae']
    df['naive_sigmav'] = df['E']/df['V'] * \
        df['sigmae'] + df['F']/df['V']*df['sigmad']
    df['DD_naive'] = (np.log(df['V']/df['F'])+(df['annret'] -
                      pow(df['naive_sigmav'], 2)/2*T))/(df['naive_sigmav']*pow(T, 0.5))
    df['PD_naive'] = norm.cdf(-df['DD_naive'])
    data['DD'] = df['DD_naive']
    data['PD'] = df['PD_naive']
    return df
DD_PD = KMV(annual, 0.05, 0.25)
# credit spread, term spread
annual = annual.merge(creditSpread, on=['year'], how='left').merge(termSpread, on = ['year'], how='left')
# remove outliers
annual.rename(columns={'sigmae': 'volatility'}, inplace=True)
annual = annual.drop(columns=['E','F'])
annual.replace([np.inf, -np.inf], np.nan, inplace=True)
annual.set_index(['year', 'cusip'], inplace=True)
annual = annual.dropna()
for col in annual.columns: 
     annual[col] = stats.mstats.winsorize(annual[col], limits=0.005)
annual
```

#### Plot ROC curve and Calculate AUC and KS statistics

```python
def rocurve1(model, X_train, X_test, y_train, y_test, y_pred):
    X = X_train.append(X_test)
    y = y_train.append(y_test)
    if isinstance(model, LogisticRegression):
        y_test_score = model.decision_function(X_test)
        y_train_score = model.decision_function(X_train)
        test_fpr, test_tpr, test_thresholds = roc_curve(y_test, y_test_score)
        train_fpr, train_tpr, train_thresholds = roc_curve(y_train, y_train_score)
        y_score = model.decision_function(X)
        fpr, tpr, thresholds = roc_curve(y, y_score)
    elif isinstance(model, RandomSurvivalForest):
        y_test_score = model.predict(X_test)
        y_train_score = model.predict(X_train)
        test_fpr, test_tpr, test_thresholds = roc_curve(y_test, y_test_score)
        train_fpr, train_tpr, train_thresholds = roc_curve(y_train, y_train_score)
        y_score = model.predict(X)
        fpr, tpr, thresholds = roc_curve(y, y_score)
    elif  isinstance(model, XGBClassifier) or isinstance(model, LGBMClassifier) or isinstance(model.estimator, KNeighborsClassifier) or isinstance(model.estimator, RandomForestClassifier):
        y_test_score = model.predict_proba(X_test)[:, 1]
        y_train_score = model.predict_proba(X_train)[:, 1]
        test_fpr, test_tpr, test_thresholds = roc_curve(y_test, y_test_score)
        train_fpr, train_tpr, train_thresholds = roc_curve(y_train, y_train_score)
        y_score = model.predict_proba(X)[:, 1]
        fpr, tpr, thresholds = roc_curve(y, y_score)
    print('Test AUC: {}'.format(auc(test_fpr, test_tpr)))
    print('Train AUC: {}'.format(auc(train_fpr, train_tpr)))
    print('Test KS: {}'.format(scipy.stats.ks_2samp(y_test, y_pred)))
    plt.figure(figsize=(10,8))
    lw = 2
    plt.plot(test_fpr, test_tpr, color='darkorange',lw=lw, label='Test ROC curve')
    plt.plot(train_fpr, train_tpr, color='blue',lw=lw, label='Train ROC curve')
    plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.yticks([i/20.0 for i in range(21)])
    plt.xticks([i/20.0 for i in range(21)])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic (ROC) Curve: X_train, X_test')
    plt.legend(loc="lower right")
    plt.show()
    # Plot tpr vs 1-fpr
    fig, ax = pl.subplots(figsize=(10,10))
    pl.plot(tpr)
    pl.plot(1-fpr, color = 'red')
    pl.xlabel('1-False Positive Rate')
    pl.ylabel('True Positive Rate')
    pl.title('Receiver operating characteristic (ROC) Curve: X')
    ax.set_xticklabels([])
    optimal_idx = np.argmax(tpr - fpr)
    mis = fpr[optimal_idx]
    print('all data cutoff misclassification rate: ', mis)
    optimal_idx = np.argmax(train_tpr - train_fpr)
    mis_train = train_fpr[optimal_idx]
    print('train data cutoff misclassification rate: ', mis_train)
    optimal_idx = np.argmax(test_tpr - test_fpr)
    mis_test = test_fpr[optimal_idx]
    print('test data cutoff misclassification rate: ', mis_test)
    return mis
```

### Method 2: Rolling out of sample estimation
#### Estimate & Forecast
* doing a rolling out of sample estimation (say, using 1964-1990 data to forecast for 1991, using 1964-1991 data for 1992 etc.)

```python
def rolling(model, X, y):
    predictionAll = pd.DataFrame()
    train_roc = pd.DataFrame()
    test_roc = pd.DataFrame()
    indecies = pd.DataFrame()
    for yr in range(1991,2020):
#         print(0)
        X_train = X.iloc[X.index.get_level_values('year') < yr]
        y_train = y.iloc[y.index.get_level_values('year') < yr]
        X_test= X.iloc[X.index.get_level_values('year') == yr]
        y_test = y.iloc[y.index.get_level_values('year') == yr]
        X_train = scaler.fit_transform(X_train)
        X_test = scaler.fit_transform(X_test)
        Model = model.fit(X_train,y_train)
        y_pred = Model.predict(X_test)
        Prob = Model.predict_proba(X_test)
        prob = Prob[:,0]
        if isinstance(model, LogisticRegression):
            y_test_score = Model.decision_function(X_test)
            y_train_score = Model.decision_function(X_train)
        elif isinstance(model.estimator, KNeighborsClassifier):
            y_test_score = Prob[:, 1]
            y_train_score = Model.predict_proba(X_train)[:, 1]
            best = model.best_estimator_
            index = pd.DataFrame(best.kneighbors(X_test, return_distance=False))
            print('for year: ', yr)
            print('best k-neighbors parameter: ', best.n_neighbors)
            print('shape of top-k index: ', index.shape)
            indecies = pd.concat([indecies, index], axis = 0)
        df_merged = pd.DataFrame({'y_test': y_test, 'y_predict': y_pred, 'no_default_prob': prob }, columns = ['y_test','y_predict', 'no_default_prob'])
        predictionAll = pd.concat([predictionAll, df_merged], axis = 0)
        train_roc_merged = pd.DataFrame({'y_train': y_train, 'y_train_score': y_train_score}, columns = ['y_train','y_train_score'])
        train_roc = pd.concat([train_roc, train_roc_merged], axis = 0)
        test_roc_merged = pd.DataFrame({'y_test': y_test, 'y_test_score': y_test_score}, columns = ['y_test','y_test_score'])
        test_roc = pd.concat([test_roc, test_roc_merged], axis = 0)
    if isinstance(model, LogisticRegression):
        return predictionAll, train_roc, test_roc
    elif isinstance(model.estimator, KNeighborsClassifier):
        index.reset_index(drop=True, inplace=True)
        return Model, indecies, predictionAll, train_roc, test_roc
```

```python
def fixed_window(model, X, y):
    predictionAll = pd.DataFrame()
    train_roc = pd.DataFrame()
    test_roc = pd.DataFrame()
    indecies = pd.DataFrame()
    for yr in range(1991,2020):
#         print(0)
        X_train = X.iloc[(X.index.get_level_values('year') >= yr-27) & (X.index.get_level_values('year') < yr)]
        y_train = y.iloc[(y.index.get_level_values('year')>= yr-27) & (y.index.get_level_values('year') < yr)]
        X_test= X.iloc[X.index.get_level_values('year') == yr]
        y_test = y.iloc[y.index.get_level_values('year') == yr]
        X_train = scaler.fit_transform(X_train)
        X_test = scaler.fit_transform(X_test)
        Model = model.fit(X_train,y_train)
        y_pred = Model.predict(X_test)
        Prob = Model.predict_proba(X_test)
        prob = Prob[:,0]
        if isinstance(model, LogisticRegression):
            y_test_score = Model.decision_function(X_test)
            y_train_score = Model.decision_function(X_train)
        elif isinstance(model.estimator, KNeighborsClassifier):
            y_test_score = Prob[:, 1]
            y_train_score = Model.predict_proba(X_train)[:, 1]
            best = model.best_estimator_
            index = pd.DataFrame(best.kneighbors(X_test, return_distance=False))
            print('for year: ', yr)
            print('best k-neighbors parameter: ', best.n_neighbors)
            print('shape of top-k index: ', index.shape)
            indecies = pd.concat([indecies, index], axis = 0)
        df_merged = pd.DataFrame({'y_test': y_test, 'y_predict': y_pred, 'no_default_prob': prob }, columns = ['y_test','y_predict', 'no_default_prob'])
        predictionAll = pd.concat([predictionAll, df_merged], axis = 0)
        train_roc_merged = pd.DataFrame({'y_train': y_train, 'y_train_score': y_train_score}, columns = ['y_train','y_train_score'])
        train_roc = pd.concat([train_roc, train_roc_merged], axis = 0)
        test_roc_merged = pd.DataFrame({'y_test': y_test, 'y_test_score': y_test_score}, columns = ['y_test','y_test_score'])
        test_roc = pd.concat([test_roc, test_roc_merged], axis = 0)
    if isinstance(model, LogisticRegression):
        return predictionAll, train_roc, test_roc
    elif isinstance(model.estimator, KNeighborsClassifier):
        index.reset_index(drop=True, inplace=True)
        return Model, indecies, predictionAll, train_roc, test_roc
```

# Model-6: Survival Random Forest
* A Survival Random Forest ensures that individual trees are not correlated. To achieve that it builds each tree on a different bootstrap sample of training data, and at each node, it only evaluates the split criterion for a randomly selected subset of features and thresholds. Follow these steps
## Out-of-sample prediction
1. Install package scikit-survival
2. Import RandomSurvivalForest() function from sksurv.ensemble} module of scikit-survival package
3. Calibrate the hyperparameter number of trees in Survival Random Forest algorithm to find the optimal value which minimizes the misclassification rate on the out-of-sample data
### Estimate & Forecast
4. Run the Survival Random Forest classification model for out-of-sample prediction

```python
from sksurv.ensemble import RandomSurvivalForest
from sksurv.metrics import (
    as_concordance_index_ipcw_scorer,
    as_cumulative_dynamic_auc_scorer,
    as_integrated_brier_score_scorer,
)
```

```python
# preprocess data
def survival_preprocess(survival_train, survival_test):
    X_train_s = survival_train.loc[:,~survival_train.columns.isin(['check_bankruptcy','survival_time'])]
    X_test_s = survival_test.loc[:,~survival_test.columns.isin(['check_bankruptcy','survival_time'])]
    y_train_s = survival_train.loc[:,survival_train.columns.isin(['check_bankruptcy','survival_time'])]
    y_test_s = survival_test.loc[:,survival_test.columns.isin(['check_bankruptcy','survival_time'])]
    # convert y to a particular array type
    y_train_s = np.array(y_train_s)
    y_test_s = np.array(y_test_s)
    dt = dtype =[('Status', '?'), ('Survival_in_days', '<f8')]
    y_train_s = np.array([tuple(row) for row in y_train_s], dtype=dt)
    y_test_s = np.array([tuple(row) for row in y_test_s], dtype=dt)
    return X_train_s, y_train_s, X_test_s, y_test_s
survival = annual.copy()
survival['survival_time'] = 1 - survival['check_bankruptcy']
survival['survival_time']= survival.groupby(by = ['cusip'])['survival_time'].transform('sum')
survival_train = survival.iloc[survival.index.get_level_values('year') <= 1990]
survival_test = survival.iloc[survival.index.get_level_values('year') > 1990]
X_train_s, y_train_s, X_test_s, y_test_s = survival_preprocess(survival_train, survival_test)
rsf = RandomSurvivalForest(verbose=3,max_depth=20,n_jobs=-1)
# fit model
rsf_param_grid = {'estimator__n_estimators':[8,16,20]}
rsf_model = GridSearchCV(as_concordance_index_ipcw_scorer(rsf), param_grid=rsf_param_grid, cv=2, n_jobs=-1)
rsf_model.fit(X_train_s, y_train_s)
print("Best hyperparameters: %f using %s" % (rfc_model.best_score_, rsf_model.best_params_))
# print("Best hyperparameters: using %s" % ('{\'estimator__n_estimators\': 20}'))
rsf = RandomSurvivalForest(rsf_model.best_params_['estimator__n_estimators'], max_depth=20, n_jobs=-1, verbose=3)
# rsf = RandomSurvivalForest(n_estimators = 20, max_depth=20, n_jobs=-1, verbose=3)
rsf.fit(X_train_s, y_train_s)
y_pred = rsf.predict(X_test_s)
print('Classify each observation in the out of sample data: y predict value \n', y_pred)
```

# Model-7: Gradient Boosted Trees (XGBOOST)
* Gradient boosting trees are the most important alternative class of tree-based estimators to random forests. Boosting is a strategy that sequentially fits an estimator to a training dataset and gives more weight to misclassified observations (or errors) in successive boosting rounds. Boosted trees iteratively estimate a sequence of shallow trees, each trained to predict the residuals of the previous tree. The final prediction is an accuracy-weighted average of the estimators. Based on the practice from previous steps, do a out-of-sample prediction using XGBoost algorithm. Follow these steps
