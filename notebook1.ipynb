# II. Method 1: KMV model
Computing Distance Default: Naive Computation using the KMV model (Method 1)



## 2.1 Compute DD PD

```python
def KMV(data, a, b):
    df = data.copy()
    T = 1
    df['V'] = df['E'] + df['F']
    df['sigmad'] = a+b*df['sigmae']
    df['naive_sigmav'] = df['E']/df['V']*df['sigmae'] + df['F']/df['V']*df['sigmad']
    df['DD_naive'] = (np.log(df['V']/df['F'])+(df['annret']-pow(df['naive_sigmav'],2)/2*T))/(df['naive_sigmav']*pow(T,0.5))
    df['PD_naive'] = norm.cdf(-df['DD_naive'])
    data['DD_naive'] = df['DD_naive']
    data['PD_naive'] = df['PD_naive']
    data.rename(columns={'DD_naive': 'DD_naive' + '_' + str(a) + '_' + str(b), 'PD_naive': 'PD_naive' + '_' + str(a) + '_' + str(b)}, inplace=True)
    return df
```

```python
KMV1 = KMV(dsf_funda_1000, 0.05, 0.25)
KMV2 = KMV(dsf_funda_1000, 0.05, 0.50)
KMV3 = KMV(dsf_funda_1000, 0, 0.25)
dsf_funda_1000
```

# III. Method 2: Direct Solve


## 3.1 Compute DD PD

The start point of sigmav and V used the solution calculated by method 1 (KMV1 = DD_naive_0.05_0.25)

```python
def fun(x, df):
    T = 1
    x1 = x[0] # x1 = V
    x2 = x[1] # x2 = sigmav
    d1 = (np.log(x1/df['F'])+(df['rf']+pow(x2,2)/2)*T)/(x2*pow(T,0.5))
    d2 = d1 - x2*pow(T,0.5)
    f = np.zeros(2)
    f[0] = df['E'] - x1*norm.cdf(d1) + np.exp(-df['rf']*T)*df['F']*norm.cdf(d2)
    f[1] = df['sigmae']- x1*x2*norm.cdf(d1)/df['E']
    return f
def direct_solve(data):
    T = 1
    df = data.copy()
    df['BS_V'], df['BS_sigmav'] =  zip(*KMV1.apply(lambda row: fsolve(fun, x0=[row['V'], row['naive_sigmav']], args=row), axis=1))
    data['DD_BS'] = (np.log(df['BS_V']/df['F'])+(df['annret']-pow(df['BS_sigmav'],2)/2*T))/(df['BS_sigmav']*pow(T,0.5))
    data['PD_BS'] = norm.cdf(-data['DD_BS'])
    return df
direct_solve(dsf_funda_1000)
dsf_funda_1000
```

# IV. Method 3 Iterative Method 


## 4.1 Compute DD PD

1. Maximum iteration is 10 and I set the epsilon as 0.001. 
2. At the same time, if the result doesn't converge and oscillate at two different values. For example, I found a sigmav point converge to two values 0.03 and 3.42. I will choose 0.03 value instead of 3.42 because 3.42 is unreal. 

```python
def fun2(x, df):
    T = 1
    d1 = (np.log(x/df['F'])+(df['rf']+pow(df['sigmav'],2)/2)*T)/(df['sigmav']*pow(T,0.5))
    d2 = d1 - df['sigmav']*pow(T,0.5)
    f = df['E'] - x*norm.cdf(d1) + np.exp(-df['rf']*T)*df['F']*norm.cdf(d2)
    return f
def iter_method(data):
    T = 1
    # initialize sigmav
    data['sigmav'] = data['sigmae']
    df = data.copy()
    df['Ite_V'] = df['F']+df['E']
    df[['converge']] = False
    # stop condition 
    epsilon = 0.005
    max_iter = 10
    i = 0
    for i in range(max_iter):
        print(i)
        if i >= 1:
#             df['converge'] = df.apply(lambda row: abs(row['Ite_sigmav']-row['sigmav'])<epsilon or (row['Ite_sigmav'] < 1 and row['sigmav'] > 3.5), axis=1)
            df['converge'] = (df['Ite_sigmav'] - df['sigmav'] < epsilon )|(df['Ite_sigmav'] < 1) & (df['sigmav'] > 3.5)
            df['sigmav'] = df['Ite_sigmav']
            if all(df['converge'] == True): 
                break 
#         print('calculate V________')
        df['Ite_V'] =  df.apply(lambda row: fsolve(fun2, x0=row['Ite_V'], args=row) if row['converge'] == False else row['Ite_V'], axis=1).astype('float')
        df['Ite_ret'] = df.sort_values(by=['date'], ascending=True).groupby(by=['cusip'])['Ite_V'].pct_change()
        df['Ite_sigmav'] = df.groupby(by=['cusip', 'year'])['Ite_ret'].transform('std')*np.sqrt(250)
#         print(df.groupby(by=['cusip', 'year'])['Ite_ret'].std()*np.sqrt(250))
    out = df.groupby(['cusip','year'], as_index=False).agg({'Ite_V':'mean', 'Ite_sigmav':'mean', 'F':'mean'})
    annret = df.groupby(by=['cusip','year']).apply(lambda x: np.exp(np.sum(np.log(1+x['ret'])))).reset_index()
    out = pd.DataFrame(annret).merge(out, on=['cusip','year'])
    out.rename(columns={0: 'annret', 'ret': 'sigmae'}, inplace=True)
    out['DD_iter'] = (np.log(out['Ite_V']/out['F'])+(out['annret']-pow(out['Ite_sigmav'],2)/2*T))/(out['Ite_sigmav']*pow(T,0.5))
    out['PD_iter'] = norm.cdf(-out['DD_iter'])
    return out
method3 = iter_method(dsf_funda_250)
method3
```
